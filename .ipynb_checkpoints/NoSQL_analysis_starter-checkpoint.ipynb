### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"Number of establishments with a hygiene score of 20: {count}")

# Display the first document in the results using pprint
result = establishments.find(query)
pprint(result[0])
```

```python
# Convert the result to a Pandas DataFrame
import pandas as pd
hygiene_df = pd.DataFrame(list(result))
# Display the number of rows in the DataFrame
print(f"Number of rows in DataFrame: {hygiene_df.shape[0]}")
# Display the first 10 rows of the DataFrame
hygiene_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {
    "LocalAuthorityName": {"$regex": "London", "$options": "i"},
    "RatingValue": {"$gte": 4}
}
# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"Number of establishments in London with a RatingValue >= 4: {count}")

# Display the first document in the results using pprint
result = establishments.find(query)
pprint(result[0])
```

```python
# Convert the result to a Pandas DataFrame
london_rating_df = pd.DataFrame(list(result))
# Display the number of rows in the DataFrame
print(f"Number of rows in DataFrame: {london_rating_df.shape[0]}")
# Display the first 10 rows of the DataFrame
london_rating_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Check if Penang Flavours was inserted correctly
penang_flavours = establishments.find_one({"BusinessName": "Penang Flavours"})

if penang_flavours:
    print("Penang Flavours found in the database:")
    pprint(penang_flavours)
else:
    print("Penang Flavours not found. Check the insertion process.")
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
degree_search = 0.01
latitude = penang_flavours["geocode"]["latitude"]
longitude = penang_flavours["geocode"]["longitude"]

query = {
    "RatingValue": 5,
    "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
    "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search}
}
sort = sort = [("scores.Hygiene", 1)]
limit = 5

# Print the results
result = establishments.find(query).sort(sort).limit(limit)
for doc in result:
    pprint(doc)
```

```python
# Convert result to Pandas DataFrame
result = list(establishments.find(query).sort(sort).limit(limit))
top_5_df = pd.DataFrame(result)
top_5_df.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Find potential duplicates based on BusinessName and LocalAuthorityName
pipeline_duplicates = [
    {"$match": {"scores.Hygiene": 0}},  # Match only those with a hygiene score of 0
    {"$group": {
        "_id": {"BusinessName": "$BusinessName", "LocalAuthorityName": "$LocalAuthorityName"},
        "count": {"$sum": 1},
        "ids": {"$push": "$_id"}
    }},
    {"$match": {"count": {"$gt": 1}}}  # Only include groups that have more than one record
]
duplicates = list(establishments.aggregate(pipeline_duplicates))
print(f"Number of duplicates found: {len(duplicates)}")
if duplicates:
    for doc in duplicates:
        pprint(doc)
```

```python
# Remove duplicates from the collection
# for duplicate in duplicates:
     # ids_to_remove = duplicate['ids'][1:]  # Keep the first document, remove the rest
     # stablishments.delete_many({"_id": {"$in": ids_to_remove}})
```

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
pipeline = [
    {"$match": {"scores.Hygiene": 0}},
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}},
    {"$limit": 10}
]
# Print the number of documents in the result
hygiene_zero_agg = list(establishments.aggregate(pipeline))
# Print the first 10 results
hygiene_zero_agg
```

```python
# Convert the result to a Pandas DataFrame
hygiene_zero_df = pd.DataFrame(hygiene_zero_agg)
# Display the number of rows in the DataFrame
print(f"Number of rows in DataFrame: {hygiene_zero_df.shape[0]}")
# Display the first 10 rows of the DataFrame
hygiene_zero_df.head(5)
```
